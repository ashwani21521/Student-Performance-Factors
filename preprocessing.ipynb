```python
import pandas as pd

data = pd.read_csv('StudentPerformanceFactors.csv')

categorical_columns = data.select_dtypes(include=['object']).columns

results = {}

for column in categorical_columns:
    mode_value = data[column].mode()[0] 
    results[column] = {
        'Mode': mode_value,
        'Unique Values': data[column].nunique(),  
        'Top 5 Values': data[column].value_counts().head(5).to_dict()  # Top 5 frequent values
    }

# Display results
for column, stats in results.items():
    print(f"Column: {column}")
    print(f"  Mode: {stats['Mode']}")
    print(f"  Unique Values: {stats['Unique Values']}")
    print(f"  Top 5 Values: {stats['Top 5 Values']}\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('StudentPerformanceFactors.csv')

# Set the style for seaborn
sns.set(style="whitegrid")

# List of categorical columns to plot
categorical_columns = [
    'Parental_Involvement', 
    'Access_to_Resources', 
    'Extracurricular_Activities', 
    'Motivation_Level', 
    'Internet_Access', 
    'Family_Income', 
    'Teacher_Quality', 
    'School_Type', 
    'Peer_Influence', 
    'Learning_Disabilities', 
    'Parental_Education_Level', 
    'Distance_from_Home', 
    'Gender'
]

# Create a combined histogram for each categorical column
plt.figure(figsize=(15, 10))

for i, column in enumerate(categorical_columns):
    plt.subplot(4, 4, i + 1)  # Adjust the layout as needed (4 rows x 4 columns)
    sns.countplot(data[column], palette='viridis')
    plt.title(column)
    plt.xlabel('Count')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Step 1: Install PyOD if you haven't already
# !pip install pyod

import pandas as pd
from pyod.models.knn import KNN  # K-Nearest Neighbors model for outlier detection
from pyod.utils.data import get_outliers_inliers
import matplotlib.pyplot as plt

# Step 2: Load the dataset
data = pd.read_csv('StudentPerformanceFactors.csv')

# Step 3: Preprocess the data
# Select numerical columns for outlier detection
numerical_columns = data.select_dtypes(include=[float, int]).columns.tolist()
X = data[numerical_columns]

# Step 4: Detect Outliers using KNN
model = KNN()
model.fit(X)

# Get predictions (1 for outlier, 0 for inlier)
y_pred = model.labels_  # Binary labels for outliers

# Add predictions to original DataFrame
data['Outlier'] = y_pred

# Step 5: Visualize results (optional)
plt.figure(figsize=(10, 6))
for column in numerical_columns:
    plt.subplot(3, 3, numerical_columns.index(column) + 1)
    plt.scatter(data.index, data[column], c=data['Outlier'], cmap='coolwarm', label='Outlier' if column == numerical_columns[0] else "")
    plt.title(column)
    plt.xlabel('Index')
    plt.ylabel(column)
plt.tight_layout()
plt.legend()
plt.show()

# Print number of detected outliers
print(f"Number of detected outliers: {y_pred.sum()}")
```
